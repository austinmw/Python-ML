# Pandas groupby quick tutorial
https://www.youtube.com/watch?v=qy0fDqoMJx8


When to use groupby? Any time you want to analyze a panda series by some category
Ex. Can rephrase question as "for each", e.g. "for each continent, what is..."


#### (Copy and paste of drinks csv data at bottom of notebook in case web link is lost)

```python
import pandas as pd
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
#drinks = pd.read_csv("drinks")
drinks.head()
```

### Question: What is the average beer servings across all countries?

```python
# simple
drinks.beer_servings.mean()
```

### Question: What if you want to look at beer servings by continent?

#### This is where a groupby is used

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
# How it works? Example:
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

# .agg()

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

## without specifying column

```python
drinks.groupby('continent').mean()
```

```python
drinks.groupby('continent').agg(['count','min','max','mean'])
```

## Easy visualization

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
